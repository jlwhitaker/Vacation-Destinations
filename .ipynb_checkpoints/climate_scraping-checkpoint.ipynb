# CLIMATE SCRAPE

```python
# URL of page to be scraped
url = 'https://www.usclimatedata.com/'

browser.visit(url)


# Retrieve page
html = browser.html

# Create BeautifulSoup object; parse with 'html'
soup = BeautifulSoup(html, 'html.parser')

print(soup.prettify())
```

```python
# Examine the results, then determine element that contains sought info
# results are returned as an iterable list
results = soup.find_all('div', class_="state_name")
results


    
```

```python
station_results = []
for result in results:
    print(result.text)
    print(result.a['href'])
    #visit the state page
    browser.visit(url+result.a['href'])
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser') 
    response = soup.find_all('div', class_="state_name")
    station_results.append(response)
    
    #get state info
    print('----------')
```

```python
print(station_results)
```

```python
#FOR SCRAPE OF ONLY PLACES IN TOP 50 DEST TABLE 
nearby_stations = []
cities = top_cities_df['City Name'].tolist()
counties = top_cities_df['County'].tolist()

#PULL NEW LIST OF ONLY TOP 50 LINKS FROM STATION RESULTS
for list in station_results:
    for result in list:
        if result.text in cities or result.text in counties:
            nearby_stations.append(result)
    
    
nearby_stations
```
